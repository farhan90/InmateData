```python

import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
from scipy.cluster.vq import kmeans,vq
from scipy.spatial.distance import cdist
from matplotlib import pyplot

df = pd.read_csv("list.csv")
x=df[['S1Q3C_1 - DESCRIBE RACE ','S1Q3C_2 - DESCRIBE RACE ','S1Q3C_3 - DESCRIBE RACE ','S1Q3C_4 - DESCRIBE RACE ','S1Q3C_5 - DESCRIBE RACE ','S1Q3C_6 - DESCRIBE RACE '] ]


x=x.replace(99,0)

x['Race']=x['S1Q3C_1 - DESCRIBE RACE ']+x['S1Q3C_2 - DESCRIBE RACE ']+x['S1Q3C_3 - DESCRIBE RACE ']+x['S1Q3C_4 - DESCRIBE RACE ']+x['S1Q3C_5 - DESCRIBE RACE ']+x['S1Q3C_6 - DESCRIBE RACE ']

x=x[['Race']]
x['Education Level']=df['S7Q1A - HIGHEST EDUCATION LEVEL PRIOR TO ADMISSION ']

x['Received GED']= df['S7Q1C - RECEIVED GED ']

x['Maximum Sentence'] = df['S4Q1B_MAXYR - (MAXIMUM) TOTAL YEARS OF SENTENCE ']

x['Maximum Sentence (Months)']=df['S4Q1B_MAXMO - (MAXIMUM) TOTAL MONTHS OF SENTENCE ']
x['Maximum Sentence (Days)']=df['S4Q1B_MAXDY - (MAXIMUM) TOTAL DAYS OF SENTENCE ']



X=x.as_matrix()
k=3
km = KMeans(k, init='k-means++') # initialize
km.fit(X)


labels = km.labels_
centroids = km.cluster_centers_




for i in range(k):
    ds = X[np.where(labels==i)]
    pyplot.plot(ds[:,0],ds[:,1],'o')
    lines = pyplot.plot(centroids[i,0],centroids[i,1],'kx')
    pyplot.setp(lines,ms=15.0)
    pyplot.setp(lines,mew=2.0)
pyplot.show()
```

##Distribution of Inmate Race

The racial distribution of inmate population in the jail. From the data we can see that White and Afrincan American races are the majority.

```python
x['Hispanic,Latino or Spanish']=df['S1Q3A - SPANISH,LATINO OR HISPANIC']


y=x[['Race','Hispanic,Latino or Spanish']]

for i,row in y.iterrows():
    if(row[0]==6 and row[1]==1):
        row[0]=7
        
        
def race_key(x):
    if(x==1):
        return 'White'
    elif(x==2):
        return 'Black or African American'
    elif(x==3):
        return 'American Indian or Alaska Native'
    elif(x==4):
        return 'Asian'
    elif(x==5):
        return 'Native Hawaiian or other Pacific Islander'
    elif(x==7):
        return 'Hispanic,Latino or Spanish'
    else:
        return 'Other'

y['Race']=y['Race'].apply(race_key)
data=y['Race'].tolist()


s = pd.Series(dict((x, data.count(x)) for x in data))
ax=s.plot(kind='bar')
ax.set_ylabel('Frequency')
#Removing column
y=y.drop('Hispanic,Latino or Spanish',1)
```
